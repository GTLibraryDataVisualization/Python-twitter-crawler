```python
import time
import datetime
import re
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
import numpy as np
import pandas as pd
import nltk
#You must download the stopwords list.  If already done, you can simply close out!
nltk.download()
from nltk.corpus import stopwords
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, HoverTool, Label
from bokeh.plotting import figure
from textblob import Blobber
from bokeh.palettes import RdYlGn
from bokeh.models import ColumnDataSource, LinearColorMapper
from bokeh.palettes import Spectral6
from bokeh.transform import linear_cmap
```

```python
#This line opens a new chrome browser for us
browser = webdriver.Chrome()
#Here we are crafting a twitter search, and asking the user to input what they want to search
base = "https://twitter.com/search?q="
query = input()
#Python works in hexidecimal, so we need to convert
#special characters so that our twitter search can read it
query_refine = query.replace(" ", "%20")
query_refine = query_refine.replace("@", "%40")
query_refine = query_refine.replace("#", "%23")
url = base + query_refine
print(url)
```

```python
#Now we redirect browser to said url
browser.get(url)
time.sleep(.6) #giving browser time to return response
```

```python
#Here we are collecting the body tag from twitter, we are then scrolling down 20 times, collecting more of the body
body = browser.find_element_by_tag_name('body')
for i in range(30):
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(.3) #allow time for browser to scroll
```

```python
tweet_list = []
#The actual text in the tweet is stored in tweet-text, so we are grabbing all elements with that tag
tweets = browser.find_elements_by_class_name('tweet-text')
#For each of the tweet text classes we have found, we are going to add them to our list, and print some stuff!
for tweet in tweets:
    print(tweet.text)
    tweet_list.append(tweet.text)
print(len(tweet_list))
```

```python
#We can create a dataframe for our sentiment analysis to make it easy!
#I want to collect the tweet text, as well as the polarity and subjectivity!
sentiment_df = pd.DataFrame(columns=['tweet', "polarity", "subjectivity"])
counter = 0
#For each of our tweets, we are first removing links and special chars
for tweet in tweet_list:
    counter += 1
    #Regex expression to remove links and special characters
    tweet = " ".join(re.sub("([^a-zA-Z0-9 ,*\u2019-])", " ", str(tweet)).split()) 
    #we are not going to create a textblob object, which computes a lot of statistics on our tweet
    blob = TextBlob(tweet)#.correct() #From here we can use TextBlob to correct any misspelled words, but it is too slow!
    #from our textblob we now select our sentiment
    sentiment = blob.sentiment
    #we can then finally add it to our dataframe at the next index.
    sentiment_df = sentiment_df.append({'tweet':tweet, "polarity":sentiment.polarity, "subjectivity":sentiment.subjectivity}, ignore_index=True)
print(sentiment_df)
    
```

```python
#Let's compute some basic statistics on our Analysis!
#these first few lines take our word list, get the most common words, and remove both stop words and the actual words we 
#searched for!  What remains is the most common words and their frequency!
word_list = pd.Series(' '.join(sentiment_df['tweet']).lower().split()).value_counts()[:50].to_frame()
word_list = word_list[~word_list.index.isin(stopwords.words('english'))]
word_list = word_list[~word_list.index.isin(query.split(" "))]
print(word_list)
#Next I am going to compute basic statistics using the pandas describe command
describe_polarity = sentiment_df['polarity'].describe()
describe_subjectivity = sentiment_df['subjectivity'].describe()
print(describe_polarity)
```

```python
#here we are telling bokey we want to show our graphs inside jupyter notebook!
output_notebook()
```

```python
#here we specify all of the tools we would like present on our graph as a string
TOOLS="crosshair,pan,wheel_zoom,zoom_in,zoom_out,box_zoom,undo,redo,reset,tap,save,box_select,poly_select,lasso_select,"
#Next we specify what we want to see on hover of a data point.  The dollar sign references data we passed into our "Source"
#parameter seen below.  an @ symbol refers to an outside defined data point.
TOOLTIPS = [
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("text", "@tweet")]
#To begin setting our colorscale, I want to find the lowest and highest polarity in our dataframe.  
#If the tweets are all positive, I don't want any to come up as red, so if the lowest numbers are not very low
#or high numbers arent high, I will set it lower to move how bokeh colors our points
low_num = min(sentiment_df["polarity"])
high_num = max(sentiment_df["polarity"])
if low_num > -0.5:
    low_num = -0.5
if high_num < 0.5:
    high_num = 0.5
#I am then creating a color mapper to associate the points with colors
#notice I put in my high number as the low point, this is to flip the colorscale!  I am using a Red Yellow Green colorscale!
#I want more negative numbers to appear redder.
mapper = linear_cmap(field_name='polarity', palette=RdYlGn[5] ,low=high_num ,high=low_num)
#Collecting date and time in order to format a title for our graph.
date = datetime.datetime.now().strftime("%I:%M%p on %B %d, %Y")
title = "Sentiment Analysis of " + query + " at " + date
#I am creating a figure with the tools I want, as well as the title I want!
p = figure(tools=TOOLS,
           title=title)
#I am now creating a scatterplot of polarity vs subjectivity, setting different parameters
#size: the size of our points
#fill_alpha: how opaque or solid I want my points
#fill_color: I am passing in our mapper so that the scatter plot knows what color to associate each value with
#source: This is where our scatterplot is going to fill in our x and y data.  we can reference it with a $
p.scatter('polarity', 'subjectivity', size=15,
        fill_alpha=0.8,
        line_color=None, fill_color=mapper, source=sentiment_df)
#We are now creating our hover tool so that on hover, we can see the tweet statistics and text!
hover = HoverTool()
hover.tooltips = TOOLTIPS
p.tools.append(hover)
#Let's edit axis labels!
p.xaxis.axis_label = "Polarity"
p.yaxis.axis_label = "Subjectivity"
#Let us add descriptive labels to our data!
#We are going to add three citations to our data! first let us create strings with what we would like to annotate
#our 3 strings contain first the 5 most common words, then the average  and STD of polarity, then average and STD of subjectivity
citation_text = "Commonly Used Words: " + word_list.index[0] + ", " + word_list.index[1] + ", " + word_list.index[2] + ", " + word_list.index[3] + ", " + word_list.index[4]
citation2_text = "Average Polarity: " + str(round(describe_polarity["mean"], 2)) + ". Standard Deviation: " + str(round(describe_polarity["std"], 2))
citation3_text = "Average Subjectivity: " + str(round(describe_subjectivity["mean"], 2)) + ". Standard Deviation: " + str(round(describe_subjectivity["std"], 2))
#Now let us format these by creating Bokeh label objects in the lower left hand corner
#Notice how we offset the y's by 20 each time.  to citation one will be on top, followed by 2, then 3
#Mode CSS means our labels will move with us when we scroll! so we can always see our labels
citation = Label(x=5, y=45, x_units='screen', y_units='screen',
                 text=citation_text, render_mode='css')
citation2 = Label(x=5, y=25, x_units='screen', y_units='screen',
                 text=citation2_text, render_mode='css')
citation3 = Label(x=5, y=5, x_units='screen', y_units='screen',
                 text=citation3_text, render_mode='css')
#we can now add our labels to our graph
p.add_layout(citation)
p.add_layout(citation2)
p.add_layout(citation3)
#Finally, we will show the graph!
show(p)
```
